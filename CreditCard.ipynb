```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
transactions_by_state = df['Merchant State'].value_counts().nlargest(10)  # Допустим, мы выберем топ-10 штатов
transactions_by_state.plot(kind='area', figsize=(10, 6), title='Количество транзакций по штатам')
plt.xlabel('Штат')
plt.ylabel('Количество транзакций')
plt.show()
```

```python
transactions_by_merchant = df['Merchant Name'].value_counts().nlargest(10)  # Предположим, мы выберем топ-10 торговых точек
transactions_by_merchant.plot(kind='bar', figsize=(10, 6), title='Количество транзакций по торговым точкам')
plt.xlabel('Торговая точка')
plt.ylabel('Количество транзакций')
plt.show()
```

```python
transaction_count_by_city = df['Merchant City'].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.pie(transaction_count_by_city, labels=transaction_count_by_city.index, autopct='%1.1f%%')
plt.title('Доля количества транзакций по городам')
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
transaction_counts = df.groupby('Month').size()
plt.figure(figsize=(10, 6))
plt.bar(months, transaction_counts)
plt.title('Количество транзакций по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество транзакций')
plt.show()
```

```python
payment_methods = df['Use Chip'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(payment_methods, labels=payment_methods.index, autopct='%1.1f%%')
plt.title('Использование методов оплаты')
plt.show()
```

```python
df['Hour'] = df['Time'].apply(lambda x: int(x.split(':')[0]))  # Извлечение часа из времени
plt.figure(figsize=(10, 6))
plt.hist(df['Hour'], bins=24, alpha=0.7)
plt.title('Распределение времени транзакций')
plt.xlabel('Время')
plt.ylabel('Количество транзакций')
plt.show()

```
